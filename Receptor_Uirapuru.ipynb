```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_S21(df, my_string):
    # Plot 'freq[Hz]' vs 'Trc1_S21[dB]'
    plt.figure(figsize=(6, 4))
    plt.plot(df['freq[Hz]'], df['Trc1_S21[dB]'], label='S21')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('S21 (dB)')
    plt.title(my_string + ' - Frequency vs S21')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def plot_S11(df, my_string):
    plt.figure(figsize=(6, 4))
    plt.plot(df['freq[Hz]'], df['Trc1_S11[dB]'], label='S11')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('S11 (dB)')
    plt.title(my_string + ' - Frequency vs S11')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def general_plot_S21(ax, df, my_string):
    ax.plot(df['freq[Hz]'], df['Trc1_S21[dB]'], label='S21')
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('S21 (dB)')
    if ax == axs[0,1]:
        ax.set_ylim(20, 35)
    if ax == axs[1,1]:
        ax.set_ylim(-60, 40)
    ax.set_title(my_string)
    ax.grid(True)
    ax.legend()
```
